```python
#Kolay Seviye Adam Asmaca

print("Adam Asmaca Oyununa Hoşgeldiniz! ")
print("Seviye = Kolay! ")

from random import randint

adamCan = 10

kelimeler = ["akbaba", "balık", "kiraz", "balina", "acemi"]
kelimeSayisi = len(kelimeler)
secilen = randint(0, kelimeSayisi-1)
secilenKelime = kelimeler[secilen]

dizilenKelime = []
for diz in secilenKelime:
    dizilenKelime.append("_")
print(dizilenKelime)

while adamCan > 0:
    girilenHarf = input("Bir harf giriniz: ")
    canKontrol = girilenHarf in secilenKelime
    if canKontrol == False:
        adamCan-=1
    i = 0
    for kontrol in secilenKelime:
        if secilenKelime[i] == girilenHarf:
            dizilenKelime[i] = girilenHarf
        i+=1
    print(dizilenKelime)
    print("Kalan can: "+ str(adamCan))
    
#Orta Seviye Adam Asmaca

print("Adam Asmaca Oyununa Hoşgeldiniz! ")
print("Seviye = Orta! ")

from random import randint

adamCan = 5

kelimeler = ["seçilme", "camekan", "ren geyiği", "uçurtma", "haberci"]
kelimeSayisi = len(kelimeler)
secilen = randint(0, kelimeSayisi-1)
secilenKelime = kelimeler[secilen]

dizilenKelime = []
for diz in secilenKelime:
    dizilenKelime.append("_")
print(dizilenKelime)

while adamCan > 0:
    girilenHarf = input("Bir harf giriniz: ")
    canKontrol = girilenHarf in secilenKelime
    if canKontrol == False:
        adamCan-=1
    i = 0
    for kontrol in secilenKelime:
        if secilenKelime[i] == girilenHarf:
            dizilenKelime[i] = girilenHarf
        i+=1
    print(dizilenKelime)
    print("Kalan can: "+ str(adamCan))

#Zor Seviye Adam Asmaca

print("Adam Asmaca Oyununa Hoşgeldiniz! ")
print("Seviye = Zor! ")

from random import randint

adamCan = 3

kelimeler = ["bukalemun", "kırlangıç", "adlandırma", "felektenbirgece", "ihtiyarlarayeryok"]
kelimeSayisi = len(kelimeler)
secilen = randint(0, kelimeSayisi-1)
secilenKelime = kelimeler[secilen]

dizilenKelime = []
for diz in secilenKelime:
    dizilenKelime.append("_")
print(dizilenKelime)

while adamCan > 0:
    girilenHarf = input("Bir harf giriniz: ")
    canKontrol = girilenHarf in secilenKelime
    if canKontrol == False:
        adamCan-=1
    i = 0
    for kontrol in secilenKelime:
        if secilenKelime[i] == girilenHarf:
            dizilenKelime[i] = girilenHarf
        i+=1
    print(dizilenKelime)
    print("Kalan can: "+ str(adamCan))
```
